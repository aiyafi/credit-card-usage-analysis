Dapat dilihat pada heatmap tersebut terdapat bagian merah yang menandakan adanya missing value yang semakin tebal berarti semakin banyak missing value-nya.

````
# Menampilkan histogram pada Credit_Limit
hist(data$Total_Trans_Amt, col='blue')
````

Pada histogram tersebut, frekuensi total transaksi selama 12 bulan terakhir memiliki frekuensi tertinggi pada total transaksi dibawah **5000**.

````
# Menampilkan histogram pada Customer_Age
hist(data$Customer_Age, col='blue')
````

Pada histogram tersebut, umur nasabah didominasi dengan yang berumur **45-50 tahun** dan paling sedikit dengan umur pada **65-70 tahun**.

````
# Menampilkan boxplot pada data Total_Trans_Ct
boxplot(data$Total_Trans_Ct, col='#1ABC9C')
````

Pada boxplot tersebut, data ```Total_Trans_Ct``` memiliki,

1. Nilai minimum berada pada angka **10**,
2. Q1 (Kuartil 1) berada pada angka **45**,
3. Q2 (Kuartil 2/Median) berada pada angka **67**,
4. Q3 (Kuartil 3) berada pada angka **81**,
5. Nilai maksimum berada pada angka **139**.

Dan pada data tersebut terdapat **dua** pencilan.

````
# Plot data dengan densitas (kepadatannya)
ggplot(data, aes(x = Total_Trans_Amt, y = Customer_Age)) +
  geom_point(alpha = .2) +
  geom_density2d()
````

Pada data diatas, diperoleh kepadatannya dengan rata-rata total dari transaksi di bawah **1000** pada rata-rata umur **15 sampai 65 tahun**.

````
# Menampilkan plot pada data Total_Trans_Amt dengan type s
plot(data$Total_Trans_Amt, type="s", col='pink')
````

Di dalam data Total jumlah transaksi bisa dilihat semakin ke kanan semakin tinggi angkanya.

````
# Menampilkan Scatter Plot pada data Credit_Limit dan Avg_Open_To_Buy
plot(x=data$Credit_Limit, y=data$Avg_Open_To_Buy, type="p", col='#2C3E50')
````

Pada scatter plot tersebut, data ```Credit_Limit``` dan ```Avg_Open_To_Buy``` membentuk pola menyerupai garis lurus semakin ke kanan, semakin ke atas, mengindikasikan bahwa ada hubungan yang erat antara ```Credit_Limit``` dan ```Avg_Open_To_Buy``` yaitu memiliki hubungan yang positif atau **(Positive Correlation)**.

````
# Menampilkan nilai korelasi pada data Credit_Limit dengan Avg_Open_To_Buy
cor(data$Credit_Limit, data$Avg_Open_To_Buy,method='pearson')
````

Dari data diatas dapat kita ketahui bahwa, koefisien korelasi dari data ```Credit_Limit``` terhadap ```Avg_Open_To_Buy``` yaitu **0.995980543932005** yang berarti terdapat hubungan yang kuat dan searah (positif) antara *Credit Limit on the Credit Card* dengan *Open to Buy Credit Line*.


````
# Digunakan untuk mengetest korelasi.
cor.test(data$Credit_Limit, data$Avg_Open_To_Buy,method='pearson')
````

````
# Menampilkan Scatter Plot pada data Customer_Age dan Months_on_book
plot(x=data$Customer_Age, y=data$Months_on_book, type="p", col='#16A085')
````

Pada scatter plot tersebut, data ```Customer_Age``` dan ```Months_on_Book``` membentuk pola menyerupai garis lurus semakin ke kanan, semakin ke atas, mengindikasikan bahwa ada hubungan yang erat antara ```Customer_Age``` dan ```Months_on_Book```  yaitu memiliki hubungan yang positif atau **(Positive Correlation)**.

````
# Menampilkan nilai korelasi pada data Customer_Age dengan Months_on_book
cor(data$Customer_Age, data$Months_on_book,method='pearson')
````

Dari data diatas dapat kita ketahui bahwa, koefisien korelasi dari data ```Customer_Age``` terhadap ```Months_on_book``` yaitu **0.788912358993053** yang berarti terdapat hubungan yang kuat dan searah (positif) antara *Customer's Age* dengan *Period of relationship with the bank*.


````
# Digunakan untuk mengetest korelasi.
cor.test(data$Customer_Age, data$Months_on_book, method='pearson')
````
